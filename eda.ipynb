**<span style="font-size:18px; color: LightBlue;">Spliting the large files intp smaller chunks</span>**

```python
file_path = "../paediatric-intensive-care-database-1.1.0/SURGERY_VITAL_SIGNS.csv"


chunk_size = 500_000  # half a million rows per chunk

for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size), start=1):
    print(f"Chunk {i} loaded:", chunk.shape)
    out_file = f"bigfile_part{i}.xlsx"
    chunk.to_excel(out_file, index=False, engine="openpyxl")
    print("Saved:", out_file, chunk.shape)
```
